#### *Prime No*

```python
lower = int(input('Enter lower Range'))
upper = int(input('Enter upper range'))

for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j == 0 :
            break 
    else :
            print(i)
```

```python
num = int(input("Enter a number: "))
x = 2
if num <=1:
    print("Not Prime")
else:
    while x < num:
        if num % x == 0:
            print("Not Prime")
            break 
        x += 1
    else:
        print("Prime")
```

```python
num= int(input("enter a number to check:- "))
if num<=1:
    print("Not Prime ")
else:
    for i in range(2,num):
        if num%i==0:
         print("Not  Prime ")
         break
    else:
        print('Prime')
```

```python
x=int(input("enter no:- "))
c=0
for i in range(2,x+1):
    if x%i==0:
        c+=1
if c>1:
    print("not a prime no")
else:
    print("Prime Number")
```

#### *Reverse number*

```python
n = 5873
num = n
reverse = 0

while num > 0:
    last_digit = num % 10
    reverse = reverse * 10 + last_digit
    num //= 10

print(reverse)
```

#### *Palindrome*

```python
n = 1234
num = n 
result =0
while num >0:
    id = num % 10
    result = (result*10)+id
    num= num//10

print(n==result)
```
