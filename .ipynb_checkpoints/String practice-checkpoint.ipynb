```python
# 17. saperate  alphabet or numeric and store in new varaiables.
x="pr1i3n5@*%$c34e44"
alpha=""
num=""
for i in x:
    if i.isnumeric():
        num=num+i
    elif i.isalpha():
        alpha=alpha+i
    else:
        pass
print(alpha)
print(num)
```

```python
# 25. Write a program to count the number of vowels _ string will given by user .


str=input("Enter strings :")
count=0
vowels="aeiouAEIOU"
for i in str:
    if i in vowels:
        count=count+1
print(" num of vowels :",count)
```

```python
# 29. '''find value with index number 
x=("sun","mon","Tue","wed","Thu","fri","sat")
for i in x:
    a=x.index(i)
    print(i," : Index No",a)
```

```python
# 30 find value with reverse index number.
x=("sun","mon","Tue","wed","Thu","fri","sat")
for i in x:
    a=x.index(i)-len(x)
    print(i," : Index No",a)
```
