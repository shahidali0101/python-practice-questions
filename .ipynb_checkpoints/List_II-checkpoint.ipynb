```python
# Sum all the items in a list
```

```python
def sum_items(list_item):

    sum = 0
    
    for i in list_item:
        sum += i
        
    return sum

print(sum_items([4,5,6,3,5,12]))
```

```python
def sum_items(list_item):

    mul = 1
    
    for i in list_item:
        mul *= i
        
    return mul

print(sum_items([4,5,6,3,5,12]))
```

```python
def max_value(n):
    largest = n[0]

    for i in n:
        if i > largest:
            largest = i

    return largest

n = [23,45,65,78,65,34,101,112,2,43]
print(max_value(n))
```

```python
# Write a Python program to count the number of strings from a given list of strings.
# The string length is 2 or more and the first and last characters are the same.

# Sample List : ['abc', 'xyz', 'aba', '1221']
```

```python
sample_list = ['abc', 'xyz', 'aba', '1221']

new=[]
count=0
for i in sample_list:
    if len(i)>2 and i[0]==i[-1]:
        count+=1
        new.append(i)
print(new)
print(count) 
```

```python
def matching(words):

    count=0
    new=[]
    
    for i in words:
        if len(i)>2 and i[0]==i[-1]:
            count+=1
            new.append(i)
    return count

words= ['abc', 'xyz', 'aba', '1221']
print(matching(words))
print(new)
            
```

```python
# Write a Python program to check if each number is prime in a given list of numbers. Return True if all numbers are prime otherwise False.
```

```python
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:              
            return False

    return True

x = [2, 3, 5, 7, 11]

print(all(is_prime(i) for i in x))


# all([True, True, True])   # True
# all([True, False, True])  # False
```

```python
# Write a Python program to change the position of every n-th value to the (n+1)th in a list.
```

```python
x = [0, 1, 2, 3, 4, 5]
n = 2

for i in range(n - 1, len(x) - 1, n):
    x[i], x[i + 1] = x[i + 1], x[i]

print(x)
```

```python
# Write a Python program to sort a list of nested dictionaries.
```

```python
students = [
    {'name': 'Ram', 'age': 22},
    {'name': 'Shyam', 'age': 20},
    {'name': 'Mohan', 'age': 25}
]

result = sorted(students, key=lambda x: x['age'])

print(result)
```
