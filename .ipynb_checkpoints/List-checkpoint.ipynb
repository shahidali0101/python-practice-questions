```python
'''Given a list of numbers,return a new list labeling 
   each as "high","medium","low".
   if number greater than 75 then show "HIGH"
   if number greater than 40 then show "Mediim"
   else show "LOW" '''
# x = [88,33,54,66,77,10,61,47,100,81,34,12]
```

```python
x = [88,33,54,66,77,10,61,47,100,81,34,12]

l = []
for i in x:
    if i >= 75:
        l.append("High")
    elif i >= 40 :
        l.append("Medium")
    else:
        l.append("Low")
print(l)
        
```

```python
# Print most freaquent value.
# x=[8,7,5,7,4,4,8,8,2,4,4,11,6]
```

```python
x = [8,7,5,7,4,4,8,8,2,4,4,11,6]

most_frequent = max(x, key=x.count)

print("Most Frequent Value:", most_frequent)
```

```python
x=[8,7,5,7,4,4,8,8,2]
c=0
v=0
for i in x:   
    count=x.count(i)
    if count>c:
        c=count
        v=i
print("most frequently value",v)
```

### *List comprehension*

```python
'''Write a python program to print Backword counting from 10 to 1
   and store in a list with comprihansion    '''
```

```python
y = [i for i in range(10,0,-1)]
print(y)
```

```python
'''create a list with random number from 10 to 80 and 
   insert 10 elements in list   '''
```

```python
import random as r
u = [r.randint(10,80) for i in range (10)]
print(u)
```

```python
# Print a (3,3) matrix using list comprehension -> Nested list comprehension
```

```python
# Cartesian product -> list comprehenson on 2 list together 
```

```python
l1= [2,3,4,5]
l2= [5,3,7,9]

[i*j for i in l1 for j in l2]
```

### *Zip*

```python
# write a program to add items of 2 lists indexwise 
```

```python
l1= [1,2,3,4]
l2= [-6,3,2,1]

list(zip(l1,l2))
[i+j for i,j in zip(l1,l2)]
```

```python
# Sum all the items in a list
```

```python
def sum_items(list_item):

    sum = 0
    
    for i in list_item:
        sum += i
        
    return sum

print(sum_items([4,5,6,3,5,12]))
```
