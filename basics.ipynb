### Python supports 3 categories of Data Types
#### 1. Basic Types
- Integer
- Float
- Complex
- Boolean
- String
#### 2. Container Types
- List
- Tuple
- Set
- Dictionary
#### 3. User-defined Types
- Class

```python
# INTEGER
```

```python
# integer
print(5)
print(1e309)
```

```python
# float
print(6.6)
print(1.7e308)
```

```python
# complex
print(5+5j)
```

```python
# String
print("Delhi")
print('Delhi')
print('''Delhi''')
```

```python
# List 
print([1,2,3,4,5])
```

```python
# Tuple 
print((1,2,3,4,5))
```

```python
# sets
print({1,2,3,4,5})
```

```python
# dict 
print({"Name":"shahid","Age":26,"Gender":"Male"})
```

### Variable 
- variables are  used to caontain or stored the values
- No need to define variable declaration in python (Dynamic Typing)
- variable in python can store multiple data type (Dynamic binding)

```python
# special syntax
a=5;b=3;c=11
print(a)
print(b)
print(c)
```

```python
a,b,c=5,4,3
print(a)
print(b)
print(c)
```

```python
a=b=c=23
print(a)
print(b)
print(c)
```
