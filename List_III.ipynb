```python
# Write a Python program to split a list every Nth element.
```

```python
c = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n']
n=3

c2 =[]

for i in range(n):
    c2.append(c[i::n])
print(c2)
    
    
```

```python
# Write a Python program to compute the difference between two lists.
```

```python
color1 = ["red", "orange", "green", "blue", "white"]
color2 = ["black", "yellow", "green", "blue"]

c1=[]
c2=[]

for i in color1:
    if i not in color2:
        c1.append(i)
print(c1)

for j in color2:
    if j not in color1:
        c2.append(j)

print(c2)
```

```python
# Write a Python program to remove key-value pairs from a list of dictionaries.
```

```python
students = [
    {'name': 'Ram', 'age': 22, 'city': 'Delhi'},
    {'name': 'Shyam', 'age': 20, 'city': 'Mumbai'},
    {'name': 'Mohan', 'age': 25, 'city': 'Pune'} ]

for d in students:
    d.pop('city')   # pop() removes and returns the removed item.

print(students)
```

```python
students = [
    {'name': 'Ram', 'age': 22, 'city': 'Delhi'},
    {'name': 'Shyam', 'age': 20, 'city': 'Mumbai'},
    {'name': 'Mohan', 'age': 25, 'city': 'Pune'} ]

for d in students:
    del d['city']      # del simply removes an item.
print(students)
```

```python
# Check if N-th Element Exists in List
```

```python
lst = [10, 20, 30, 40, 50]

n = int(input("Enter position: "))

if n <= len(lst):
    print("Element exists:", lst[n-1])
else:
    print("Element does not exist")
```

```python
# Write a Python program to find a tuple, the smallest second index value from a list of tuples.
```

```python
x = [(1, 5), (2, 3), (4, 8), (6, 1)]

smallest = x[0]

for i in x:
    if i[1] < smallest[1]:
        smallest = i

print(smallest)
```
