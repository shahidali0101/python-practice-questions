```python
# Write a Python script to sort (ascending and descending) a dictionary by value.
```

```python
d = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}

# Sort dictionary by value in ascending order
ascending = dict(sorted(d.items(), key=lambda x: x[1]))

# Sort dictionary by value in descending order
descending = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))

print("Original Dictionary:", d)
print("Ascending Order:", ascending)
print("Descending Order:", descending)
```

```python
# Write a Python script to add a key to a dictionary.
```

```python
s = {0: 10, 1: 20}
s[3]=30
print(s)
```

```python
# Write a Python script to concatenate the following dictionaries to create a new one.

# Sample Dictionary :
# dic1={1:10, 2:20}
# dic2={3:30, 4:40}
# dic3={5:50,6:60}
```

```python
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}

new_dic={}

new_dic.update(dic1)
new_dic.update(dic2)
new_dic.update(dic3)

print(new_dic)
```

```python
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}

new_dic={**dic1,**dic2,**dic3}     # Using Dictionary Unpacking (**)  (Python 3.5+)

print(new_dic)
```

```python
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}

new_dic={}

for d in (dic1,dic2,dic3):
    for key , values in d.items():
        new_dic[key]=values
print(new_dic)
```
