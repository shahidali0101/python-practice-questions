```python
# create a function chek number is even or odd.
def is_even(number):
    """
    This function tells if a given number is even or odd
    input - any valid integer
    Output - Odd/even 
    last edited - 21/Mar/2026
    """
    if number%2==0:
        return "Even"
    else:
        return "Odd"
        
is_even(21)  
```

```python
for i in range(1,10):
    print(i,is_even(i))
```

```python
# code improvement
def is_even(number):
    """
    This function tells if a given number is even or odd
    input - any valid integer
    Output - Odd/even 
    last edited - 21/Mar/2026
    """
    if type(number)==int:
        if number%2==0:
            return "Even"
        else:
            return "Odd"
    else:
        return("Not a valid no")
```

```python
# Create a function to cunt Even number .
def count_even(n):
    c=0
    for i in n:
        if i%2==0:
            c+=1
    print("Total Number of even :-  ",c)
n=[4,5,6,7,8,9,34,56,76,34,22,33,11]
count_even(n)
```
