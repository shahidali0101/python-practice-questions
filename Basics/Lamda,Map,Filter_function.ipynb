## *Lambda*


### *A lambda function in Python is a small, anonymous (unnamed) function defined using the lambda keyword. It’s typically used for short, one-line operations where writing a full function with def would be unnecessary.*

```python
# lamda input:expression
```

```python
# square 

a=lambda x:x**2 
a(8)
```

```python
a= lambda x,y: x+y 
a(6,7)
```

```python
b=lambda x: "Even" if x%2==0 else "Odd"
b(22)
```

### *Map*
*In Python, map() is a built-in function used to apply a function to every item in an iterable (like a list, tuple, etc.) and return a new iterator with the results.*

```python
numbers = [1, 2, 3, 4]

result = list(map(lambda x: x**2, numbers))
print(result)   # [1, 4, 9, 16]
```

```python
l=[3,4,6,7,9,11]
m=map(lambda x:x*2,l)
```

```python
list(m) 
```

```python
# check a list has the elements Even or odd
l=[3,4,6,7,9,11]
m=list(map(lambda x:x%2==0 ,l))
m
```

### *Filter*
*In Python, filter() is a built-in function used to select elements from an iterable based on a condition.*

```python
l=[3,4,6,7,9,11]
list(filter(lambda x:x>6,l))
```

```python
fruits=['Apple','Orange','Mango','Guava']
list(filter(lambda x:'e' in x,fruits))
```

### *List Comprehension*
*List Comprehension in Python is a short and powerful way to create lists in a single line.*

```python
# double of element in a list.

l=[2,4,7,5,9,11,13]
l1=[item*2 for item in l]
l1
```

```python
l2=[i**2 for i in range(10)]
l2
```

```python
# only odd numbers in a list
l3=[i**2 for i in range(11) if i%2!=0]
l3
```
