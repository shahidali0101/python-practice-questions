```python
# Sum all the items in a list
```

```python
def sum_items(list_item):

    sum = 0
    
    for i in list_item:
        sum += i
        
    return sum

print(sum_items([4,5,6,3,5,12]))
```

```python
# Write a Python program to multiply all the items in a list.
```

```python
def sum_items(list_item):

    mul = 1
    
    for i in list_item:
        mul *= i
        
    return mul

print(sum_items([4,5,6,3,5,12]))
```

```python
def max_value(n):
    largest = n[0]

    for i in n:
        if i > largest:
            largest = i

    return largest

n = [23,45,65,78,65,34,101,112,2,43]
print(max_value(n))
```

```python
# Write a Python program to count the number of strings from a given list of strings.
# The string length is 2 or more and the first and last characters are the same.

# Sample List : ['abc', 'xyz', 'aba', '1221']
```

```python
sample_list = ['abc', 'xyz', 'aba', '1221']

new=[]
count=0
for i in sample_list:
    if len(i)>2 and i[0]==i[-1]:
        count+=1
        new.append(i)
print(new)
print(count) 
```

```python
def matching(words):

    count=0
    new=[]
    
    for i in words:
        if len(i)>2 and i[0]==i[-1]:
            count+=1
            new.append(i)
    return count

words= ['abc', 'xyz', 'aba', '1221']
print(matching(words))
print(new)
            
```

```python
# Write a Python program to get a list,
# sorted in increasing order by the last element in each tuple from a given list of non-empty tuples.
# Sample List : [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
```

```python
x = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]

def last_element(t):
    return t[-1]

y = sorted(x, key=last_element)

print(y)
```

```python
# Write a Python program to print a specified list after removing the 0th, 4th and 5th elements.
# Sample List : ['Red', 'Green', 'White', 'Black', 'Pink', 'Yellow']
```

```python
x = ['Red', 'Green', 'White', 'Black', 'Pink', 'Yellow']

y = []

for i in range(len(x)):
    if i not in [0, 4, 5]:
        y.append(x[i])

print(y)
```
