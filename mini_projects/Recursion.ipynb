# *Recursion*

### *Recursion in Python means a function calls itself to solve a problem.*

```python
# multiply with loop

def multi(a,b):
    result=0
    
    for i in range(b):
        result=result+a
        
    print(result)
    
multi(3,4)
```

```python
# with recursion
def mul(a,b):
    if b==1:
        return a
    else:
        return a + mul (a,b-1)
print(mul(6,7))
```

```python
# factorial
def fc(number):
    if number==1:
        return 1
    else:
        return number * fc(number-1)
print(fc(5))
```

```python
# Palindorm
def palin(text):
    if len(text)<=1:
        print("Palindrome")
    else:
        if text[0]==text[-1]:
            palin(text[1:-1])
        else:
            print("Not a Palindrom")
palin("madam")
palin("manm")
```

```python
# fibonacci

def fib(m):
    if m==0 or m==1:
        return 1
    else:
        return fib(m-1) + fib(m-2)
print(fib(12))
```
